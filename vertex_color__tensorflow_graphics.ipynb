```
import numpy as np
#from tensorflow_graphics.geometry.transformation import quaternion
from tensorflow_graphics.notebooks import threejs_visualization
#from tensorflow_graphics.geometry.representation.mesh import utils as mesh_utils
import urllib.request
```

```
url = "https://raw.githubusercontent.com/sandroormeno/tensorflow_graphics/master/modelo.ply"
urllib.request.urlretrieve(url, 'modelo.ply')
file = open("modelo.ply", "r")
```

```
vertices = ()
faces = ()
colorvertex = ()
for line in file: 
    my_linea = line
    x = my_linea.split(" ")
    if len(x) == 12:
        vertice_tem = (float(x[0]), float(x[1]), float(x[2])),
        vertices = vertices + vertice_tem

        tem_color = (int(x[8]) , int(x[9]),  int(x[10])),
        colorvertex = colorvertex + tem_color
    if len(x) == 4:

        cara_tem = (int(x[1]), int(x[2]), int(x[3])), 
        faces = faces + cara_tem
        
vertices = np.array(vertices)
faces = np.array(faces)
colorvertex = np.array(colorvertex)
colorvertex = colorvertex/255
```

```
# https://www.htmlcsscolor.com/hex/808080

context = threejs_visualization.build_context()
light1 = context.THREE.PointLight.new_object(0XDB3838)
light1.position.set(50., 10., 10.)
#light2 = context.THREE.PointLight.new_object(0x6AA84F)
light2 = context.THREE.AmbientLight.new_object(0x6AA84F)# 0x808080
light2.position.set(-50., 10., 10.)
lights = (light1, light2)
material = context.THREE.MeshLambertMaterial.new_object( { 'color': 0xfffacd, 'vertexColors': context.THREE.VertexColors , 'side': context.THREE.DoubleSide ,} )
mesh = {'vertices': vertices, 'faces': faces,'vertex_colors' : colorvertex,   'material': material, 'castShadow' : True , 'receiveShadow' : True }
camera = threejs_visualization.build_perspective_camera(
        field_of_view=30, position=(10.0, 10.0, 10.0))
width = 500
height = 500
_ = threejs_visualization.triangular_mesh_renderer([mesh],
                                                   width=width,
                                                   height=height,
                                                   camera=camera,
                                                   lights=lights)
```
